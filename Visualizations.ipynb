```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("iris.csv")
```

```python
df.head()
```

```python
#Setting the column names
df.columns = ["sepal length","sepal width","petal length","petal width","type"]
```

## Plotting

```python
plt.figure(figsize=(20,6))
sns.violinplot(data = df)
plt.show()
```

```python
plt.figure(figsize=(20,6))
p = sns.lineplot( data = df['sepal length'])
p.set( xlabel = "sepal length", ylabel = "cm")
plt.show()
```

```python
df["type"].unique()
```

```python
#Correlation visualization
plt.figure(figsize=(20,6))
sns.heatmap(df.corr()).set_title('Correlation heat map')
plt.show()
```
